### Operators

```python
## Airthmetic Operations
a=10
b=15
add_result = a+b
print(add_result)  # 25
sub_result = a-b
print(sub_result)  # -5
mul_result = a*b
print(mul_result)  # 150
div_result = a/b
print(div_result)  # 0.6666666666666666
floor_div_result = a//b
print(floor_div_result)  # 0
modulus_result = a%b
print(modulus_result)  # 10
exponent_result=a**b
print(exponent_result)  # 100000000000000000000
```

### Logical Operators

```python
## And , Both will be true.
x=True
y=True
result = x and y
print(result)
```

```python
## Or , both are true than only false comes.
x=False
y=False
result = x or y
print(result)
```

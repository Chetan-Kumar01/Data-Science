```python
## Why functions
num = 24
if num%2 == 0:
    print(f"{num} is even") 
else:
    print(f"{num} is odd")
```

```python
def even_or_odd(num):
    """This function checks if a number is even or odd"""
    if num%2 == 0:
        return f"{num} is even"
    else:
        return f"{num} is odd"
```

```python
## Calling a function
even_or_odd(23)
```

```python
## Function with multiple parameters
def add_numbers(num1, num2):
    """This function adds two numbers"""
    return num1 + num2
```

```python
add_numbers(3, 5)
```

```python
## Default parameters
def greet(name, msg="Good morning!"):
    """This function greets to the person with the provided message"""
    print(f"Hello {name}, {msg}")
greet("Kate")
greet("Bruce", "How do you do?")
```

```python
## Variable length arguments
## Positional arguments

def student_info(*args):
    for number in args:
        print(f"Student number: {number}")

student_info('Math', 'Art', 'History')
```

```python
## Keyword arguments

def print_details(**kwargs):
    for key, value in kwargs.items(): ## In keyword arguments all the parameters will be in the form of key-value pair
        print(f"{key}: {value}")

print_details(name="John", age=22, city="New York")
```

```python
## Positional and keyword arguments
def display_info(*args, **kwargs):
 
    for arg in args:
        print(f"Positional argument- {arg}")
    for key, value in kwargs.items():
        print(f"Keyword argument- {key}: {value}")
```

```python
display_info(1, 2, 3, 4, name="Alice", age=30)
```

```python
## Return statements
def square(num):
    """This function returns the square of a number"""
    return num * num

square(3)
```

```python
## Return multiple values
def arithmetic_operations(num1, num2):
    """This function returns the sum, difference, product and quotient of two numbers"""
    sum_result = num1 + num2
    diff_result = num1 - num2
    prod_result = num1 * num2
    quot_result = num1 / num2
    return f"Sum: {sum_result}, Difference: {diff_result}, Product: {prod_result}, Quotient: {quot_result}"

arithmetic_operations(10, 2)
```

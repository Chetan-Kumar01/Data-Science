### Pandas-DataFrame and Series

Pandas is a powerful data manipulatioon library in python ,widely used for data analusis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is a one-Dimensional array like object.

```python
import pandas as pd
```

```python
## Series is a one dimensional series.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series\n ", series)
```

```python
## Create a series from dictionary
data={'a':1,'b':2,'c':3,'d':4,'e':5}
series_dictionary=pd.Series(data)
print("Series from dictionary\n ", series_dictionary)
```

```python
data = [10,20,30]
index = ['a','b','c']
series_with_index = pd.Series(data,index=index)
print("Series with custom index\n ", series_with_index)
```

```python
## Dataframe
## create a dataframe from dictionary of list
data={
    'Name' :['Krish','John','Jack'],
    'Age' :[25,26,24],
    'City' :['New York','Los Angeles','Chicago']
}

df=pd.DataFrame(data)
print(df)
```

```python
import numpy as np
np.array(df)
```

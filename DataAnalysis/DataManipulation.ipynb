```python
df_filled=df.fillna(0)
```

```python
## filling missing values with mean of column.
df['Salary_fillNA']=df['Salary'].fillna(df['Salary'].mean())
df
```

```python
df.dtypes
```

```python
## rename the columns
df=df.rename(columns={'ID':'College Id'})
df
```

```python
## change datatype
df['Age_new']=df['Age'].fillna(df['Age'].mean()).astype(int)
df
```

```python
df['Age_function']=df['Age'].apply(lambda x:x*2)
df
```

```python
## Data aggregating and grouping
df
```

```python
grouped_mean=df.groupby('Name')['Salary_fillNA'].mean()
print(grouped_mean)
```

```python
## Merging and joining Dataframes
# Create sample DataFrames
df1=pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'], 'Value2':[4,5,6]})
df1
```

```python
df2
```

```python
## Merge Dataframe 
pd.merge(df1,df2,on="Key",how="inner")  # inner, outer, left, right
```

```python
pd.merge(df1, df2, on="Key", how="left")
```

```python
pd.merge(df1, df2, on="Key", how="right")
```

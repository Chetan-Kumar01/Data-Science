### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

## create a line plot
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
## plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

##  create a customize line plot
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## Multiple plots
## Sample Data
x1=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,9,16,25]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='purple')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x1,y2,color='green')
plt.title("Plot 4")
```

```python
## bar plot
categories=['A', 'B', 'C', 'D']
values=[23,45,12,36]

## create a bar plot
plt.bar(categories, values,color='purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

### Histograms

```python
data=[23,45,12,36,78,45,67,89,23,45,12,36,78,45,67,89,23,45,12,36,78,45,67,89]

## create a histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
```

```python
## Create a scatter plot
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.scatter(x,y,color="blue",marker='o')
```

```python
## pie chart

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)

## create a pie chart
plt.pie(sizes, labels=labels, explode=explode, colors=colors, autopct="%1.1f%%", shadow=True)
```

```python
## Sales data visulization
import pandas as pd
```

```python
sales_data_df=pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
sales_data_df.info()
```

```python
## plot total sales by products
total_sales_by_product=sales_data_df.groupby('Product_Name')['Total_Revenue'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar', color='skyblue')
```

```python
## plot sales trend over time
sales_trend=sales_data_df.groupby('Date')['Total_Revenue'].sum().reset_index()
sales_trend
```

```python
sales_trend.plot(kind='line', color='purple', marker='o')
```

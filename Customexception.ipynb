### Custom exception (Raise and Throw an exception)

```python
class error(Exception):
    pass

class dobException(error):
    pass
```

```python
year=int(input("Enter your birth year: "))
age=2025-year

try:
    if age<=30 and age>=20:
        print("The age is valid to apply for the job")

    else:
        raise dobException
except dobException:
    print("The age is not valid to apply for the job")
```

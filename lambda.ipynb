### Lambda Functions In Python

Lambda functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only one expression.

They are commomnly used for short operations or as arguments to higher-order functions.

```python
# Syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b

print(addition(5,6))  # Output: 11
```

```python
addition = lambda a,b: a+b 
type(addition)  # Output: <class 'function'>
print(addition(5,6))  # Output: 11
```

```python
def even(num):
    if num%2==0:
        return True

even(24)
```

```python
even1 = lambda num: num%2==0
even1(24)
```

```python
def add(x, y, z): 
    return x + y + z

add(2, 3, 4)
```

```python
add1 = lambda x, y, z: x + y + z
add1(2, 3, 4)
```

```python
## Map() -> applies a given function to all items in an input list (or any iterable) and returns a map object (which is an iterator).
numbers=[1,2,3,4,5]
def square(num):
    return num**2

square(3)
```

```python
map(lambda x: x**2, numbers)  # Output: <map object at 0x7f9b8c0d1d90>
list(map(lambda x: x**2, numbers))  # Output: [1, 4, 9, 16, 25]
```

### Sql and Sqlite
SQL is a standard language for managing and manipulating relational databases. SQLite is a self-contained, services, and zero-configuration database engine that is widely used for enbedded tables, and performing various SQL operations.

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
### Create a Table
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT 
)
''')
## commit
connection.commit()
```

```python
cursor.execute('''
Select * from employees 
     ''')
```

```python
## Insert the data in sqlite table 
cursor.execute('''
INSERT INTO employees (name, age, department) 
   VALUES ('John Doe', 30, 'IT')
    
''')

cursor.execute('''
INSERT INTO employees (name, age, department)
   VALUES ('Jane Smith', 25, 'HR')
''')

cursor.execute('''
INSERT INTO employees (name, age, department) 
    VALUES ('Mike Johnson', 35, 'Finance')
''')
## commit
connection.commit()
```

```python
## Clear the table.
cursor.execute('DELETE FROM employees')
connection.commit()
```

```python
## Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

## print the data

for row in rows:
    print(row)
```

```python
cursor.execute('''
UPDATE employees 
   SET age = 31 
 WHERE name = 'John Doe'
''')

connection.commit()
```

```python
## Working with sales data
## Connect with SQLite data base
connection=sqlite3.connect('sales_data.db')
cursor=connection.cursor()
```

```python
## Create a table for sales data
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT
)
''')

## Insert the data in sqlite table
sales_data= [
    ('2023-10-01', 'Product A', 100, 'North'),
    ('2023-10-01', 'Product B', 150, 'South'),
    ('2023-10-02', 'Product A', 200, 'East'),
    ('2023-10-02', 'Product C', 250, 'West'),

]

## Bulk insertion.
cursor.executemany('''
INSERT INTO sales (date, product, sales, region) 
   VALUES (?, ?, ?, ?)
''', sales_data)

## Commit
connection.commit()
```

```python
## Query the data from the table.
cursor.execute('Select * from sales')
rows=cursor.fetchall()
## print the data
for row in rows:
    print(row)
```

```python
# Close the connection
connection.close()
```

### Variables

```python
## Declaring and Assigning Variables

age = 32
height = 5.10
name = "John"
is_student = False

## printing the variables
print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

```python
## Naming conventions
## Variable names should be descriptive and follow these guidelines:
## - Use lowercase letters for variable names.
## - Separate words with underscores (e.g., `first_name`, `last_name`).
## - Avoid using special characters or spaces.
## - Start variable names with a letter or underscore, not a number.

first_name="Krish"
last_name="Kumar"
```

```python
## Invalid Variable Names
#1st_name = "Krish"
# last-name = "Kumar"
#@name = "Krish"
```

```python
## case sensitivity
name = "Krish"
Name = "Kumar"

name == Name  # This will be False because variable names are case-sensitive
```

```python
## Understanding variable types
## Python is dynamically typed , type of a varaible is determined at runtime.
age=24
height=5.10 #float
name="krish" #str
print(type(age))    # <class 'int'>
print(type(height)) # <class 'float'>
print(type(name))    # <class 'str'>
```

```python
## Dynamic Typing
## Python allows variables to change type
var=5
print(var , type(var))  # var is an integer
var="Hello"
print(var , type(var))  # var is now a string\
var=3.14
print(var , type(var))  # var is now a float
```

```python
## Input

age=input("Enter your age: ") ## when we use input then it converts the input to a string.
print(type(age))    # <class 'str'>
age=int(age)
print(age ,type(age))    # <class 'int'>
```

```python
## Calculator
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")
operation = input("Enter operation (+, -, *, /): ")

# Convert inputs to float
num1 = float(num1)
num2 = float(num2)

if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    result = num1 / num2
else:
    result = "Invalid operation"

print("Result:", result)
```

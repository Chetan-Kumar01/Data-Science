### To Do List

```python
to_do_list = ["Groceries" , "Clean the house", "Pay bills"]

## Adding task
to_do_list.append("Walk the dog")
to_do_list.append("Go for a walk")

## Removing tasks
to_do_list.remove("Clean the house")

## checking if the task is in the list
if "Pay bills" in to_do_list:
    print("Don't forget to pay the bills!")
print("To do list remaining")
for task in to_do_list:
    print(f"- {task}")
```

### Calculate grades

```python
# Organizing student grades
grades = [85, 90, 78, 92, 88]

# Adding a new grade
grades.append(95)

# Calculating the average grade
average_grade = sum(grades) / len(grades)
print(f"The average grade is: {average_grade}")

# Finding the highest and lowest grade
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"The highest grade is: {highest_grade}")
print(f"The lowest grade is: {lowest_grade}")
```

### Managing the Inventory

```python
# Managing the Inventory
inventory = ["apple", "banana", "orange"]

# Adding items to the inventory
inventory.append("grape")
inventory.append("watermelon")

# Removing an item from the inventory
inventory.remove("banana")

# checking if an item is out of stock
item = "apple"
if item in inventory:
    print(f"{item} is available in the inventory.")
else:
    print(f"{item} is out of stock.")
print("Current inventory:")
for item in inventory:
    print(f"- {item}") 
```

```python
# Collecting user feedback
feedback = ["Great service!" , "Very satisfied" , "Could be better", "excellent experience"]

# Adding new feedback
feedback.append("Not happy with the service")

# counting specific feedback
positive_feedback_count = sum(1 for f in feedback if "great" in f.lower() or "satisfied" in f.lower() or "excellent" in f.lower())
negative_feedback_count = sum(1 for f in feedback if "not happy" in f.lower() or "could be better" in f.lower())
print(f"Positive Feedback Count: {positive_feedback_count}")
print(f"Negative Feedback Count: {negative_feedback_count}")

# Printing all feedback
print("User Feedback:")
for f in feedback:
    print(f"- {f}")
```

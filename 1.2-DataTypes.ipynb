### Data Types

```python
 ## Integer
age = 25
print(age, type(age))  # 25 <class 'int'>
```

```python
## Float
height = 5.9
print(height, type(height))  # 5.9 <class 'float'>
```

```python
## Bool
is_student = True
print(is_student, type(is_student))  # True <class 'bool'>

a=10
b=10
type(a==b)
```

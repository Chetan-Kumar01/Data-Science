### Python Logging

Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in-logging module offers a flexible framework for emmiting log messages from Python programs. In this lesson, we will cover the basics of logging, including hot to configure logging, log levels and best practices for using logging in python applications.

```python
import logging

## Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

## log messages
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

### Log Levels

Python's logging module defines several log levels, each representing the severity of the log message. The standard log levels are:
- DEBUG: Detailed information, typically of interest only when diagnosing problems.
- INFO: Confirmation that things are working as expected.
- WARNING: An indication that something unexpected happened, or indicative of some problem in the near future
- ERROR: Due to a more serious problem, the software has not been able to perform some function.
- CRITICAL: A very serious error, indicating that the program itself may be unable to continue


```python
## configuring logging
logging.basicConfig(
    filename='app.log',
    filemode='w',  # 'w' to overwrite, 'a' to append
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True  # this ensures reconfiguration
)

# log message
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

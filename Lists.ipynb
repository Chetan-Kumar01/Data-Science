```python
set1={10,20,30,40}
print(list(set1))
```

```python
names = ["chetu" , "janmejay" , 1, 2]
print(names)
```

```python
mix_elements = ["chetu" , "janmejay" , 1, 2, True, 5.6]
print(mix_elements)
```

```python
## Accessing elements
print(names[0])
print(mix_elements[1])
print(mix_elements[4])
# --- IGNORE ---
```

```python
print(mix_elements[1:])
print(mix_elements[1:4])
print(mix_elements[:3])
```

```python
## Modifying elements
names[0] = "new_name"
mix_elements[1] = "new_element"
print(names)
print(mix_elements)
```

```python
fruits = ["apple", "banana", "cherry"]
## Lists Methods
fruits.append("orange")  # Add an element to the end of the list
fruits.insert(1, "blueberry")  # Insert an element at a specific position
fruits.remove("banana")  # Remove an element by value
print(fruits)
```

```python
## Remove and return the last element.
last_fruit = fruits.pop()
print(last_fruit)
print(fruits)
```

```python
index = fruits.index("cherry")  # Get the index of an element
print(index)
count = fruits.count("apple")  # Count occurrences of an element
print(count)
```

```python
fruits.sort()  # Sort the list in ascending order
print(fruits)
fruits.reverse()  # Reverse the order of the list
print(fruits)
```

```python
## Slicing Lists
numbers = [1, 2, 3, 4, 5]
print(numbers[1:3])
print(numbers[:2])
print(numbers[1:])
print(numbers[-1:])  # Last element
print(numbers[-3:-1])  # Slice from the end
print(numbers[::2])  # Every second element
print(numbers[::-1])  # Reverse the list
```

```python
## Iterating Through Lists
for fruit in fruits:
    print(fruit)

for number in numbers:
    print(number)
```

```python
## Iterate with Index
for index,number in enumerate(numbers):
    print(f"Index: {index}, Number: {number}")
```

```python
## List comprehensions
squared_numbers = [x**2 for x in numbers]
print(squared_numbers)

filtered_fruits = [fruit for fruit in fruits if "a" in fruit]
print(filtered_fruits)

lst = []
for i in range(10):
    lst.append(i**2)
print(lst)
```

```python
[i**2 for i in range(10)]  # List of numbers from 0 to 9
```

### List Comprehension

Basic Syntax              [expression for item in iterable]

with conditional logic    [expression for item in iterable if condition]

Nested List Comprehension  [expression for item1 in iterable1 for item2 in iterable2]

```python
### Basic list comprehensions
squared_numbers = [x**2 for x in numbers]
print(squared_numbers)

filtered_fruits = [fruit for fruit in fruits if "a" in fruit]
print(filtered_fruits)
```

```python
even_numbers = [num for num in range(10) if num % 2 == 0]  # Even numbers from 0 to 9
print(even_numbers)
```

```python
## Nested List Comprehensions
lst1 = [1, 2, 3]
lst2 = ['a', 'b', 'c']
nested = [[i,j] for i in lst1 for j in lst2]
print(nested)
```

```python
## List Comprehensions with function calls
words = ["hello", "world", "python"]
capitalized_words = [word.upper() for word in words]
print(capitalized_words)
lengths = [len(word) for word in words]
print(lengths)
```

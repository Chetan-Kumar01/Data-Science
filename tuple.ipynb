### Tuple

```python
## Accessing tuple elements
numbers = (1, 2, 3, 4, 5)
print(numbers[0])  # Output: 1
print(numbers[1])  # Output: 2
print(numbers[2])  # Output: 3
print(numbers[3])  # Output: 4
print(numbers[4])  # Output: 5
```

```python
print(numbers[-2])
print(numbers[0:3])
print(numbers[::-1])
```

```python
## Tuple operations.
mixed_numbers = (1, "two", 3.0, (4, 5))
print(mixed_numbers[0])      # Output: 1
print(mixed_numbers[1])      # Output: "two"
print(mixed_numbers[2])      # Output: 3.0
print(mixed_numbers[3])      # Output: (4, 5)

concatenation_tuple = numbers + mixed_numbers
print(concatenation_tuple)
print(numbers * 2)
print(mixed_numbers * 2)
```

```python
numbers[1] = "krish"
```

```python
## Immutable nature of tuple
# Trying to change a tuple will result in an error
try:
    numbers[0] = 10
except TypeError as e:
    print(e)  # Output: 'tuple' object does not support item assignment
```

```python
## Tuple methods
numbers = (1, 2, 2, 3, 4, 5)
print(numbers.count(2))
print(numbers.index(3))
```

```python
## Packing and Unpacking tuples
# Packing
person = ("John", 30, "Engineer")
print(person)
# Unpacking
name, age, profession = person
print(name)
print(age)
print(profession)

## Unpacking with *
# Using * to unpack
person = ("John", 30, "Engineer", "USA")
name, age, *other_info = person
print(name)
print(age)
print(other_info)
```

```python
## Nested Tuples
nested_tuple = (1, 2, (3, 4), 5)
print(nested_tuple[0])        # Output: 1
print(nested_tuple[2])        # Output: (3, 4)
print(nested_tuple[2][0])     # Output: 3
```

```python
## Iterating over nested tuple
nested_tuple = ((1,2,3),("a","b","c"),(True,False))
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

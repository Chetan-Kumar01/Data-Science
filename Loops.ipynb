### Loops

```python
## for loop

for i in range(5):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
```

```python
## strings

str = "Chetu"
for i in str:
    print(i)
```

```python
## while loop

i = 0
while i < 5:
    print(i)
    i += 1
```

```python
count = 0
while count%2 == 0:
    print(count)
    count += 1
```

```python
## Loop control statements
## break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue statement
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
## pass statement
for i in range(10):
    if i == 5:
        pass
    print(i)
```

```python
## nested loops
for i in range(5):
    for j in range(5):
        print(i, j)
```

```python
## Examples - Calculate the sum of first n natural numbers using while loop
n = 10
sum = 0
i = 1
while i <= n:
    sum += i
    i += 1
print(sum)
```

```python
## using for loop.

n= 10
sum = 0
for i in range(1,11):
    sum += i
print(sum)
```

```python
## Prime numbers.(1,100)
## for loop
for i in range(1,101):
    if i>1:
        for j in range(2,i):
            if(i%j)==0:
                break
        else:
            print(i)
```
